```python
def countstring(x):
    x=x.lower()
    vowelcount=0
    consonantscount=0
    n=len(x)
    for i in range(n):
        if x[i] in "aeiou":
            vowelcount+=1
        else:
            consonantscount+=1
    return vowelcount,consonantscount


str=input("Enter the string:")
countstring(str)
```

```python
r1 = int(input("Enter number of rows: "))
c1 = int(input("Enter number of cols"))

r2 = int(input("Enter number of rows: "))
c2 = int(input("Enter number of cols"))

matrix1 =[]
matrix2 =[]

for i in range(r1):
  row =[]
  for j in range(c1):
    x = int(input(f"Enter the element[{i}],[{j}]: "))
    row.append(x)
  matrix1.append(row)
print(matrix1)

for i in range(r2):
  row =[]
  for j in range(c2):
    x = int(input(f"Enter the element[{i}],[{j}]: "))
    row.append(x)
  matrix2.append(row)
print(matrix2)

if(c1 != r2):
  print("error")
else:
  matrix3 = []
  for i in range(r1):
        row = []
        for j in range(c2):
            s = 0
            for k in range(c1):
                s += matrix1[i][k] * matrix2[k][j]
            row.append(s)
        matrix3.append(row)

  print(matrix3)
```

```python
size1 = int(input("Enter the size of first list: "))
list1 = []
for i in range(size1):
    val = int(input("Enter element: "))
    list1.append(val)

size2 = int(input("Enter the size of second list: "))
list2 = []
for i in range(size2):
    val = int(input("Enter element: "))
    list2.append(val)

common_count = 0
for i in range(size1):
    for j in range(size2):
        if list1[i] == list2[j]:
            common_count += 1
            break

print("Number of common elements:", common_count)
```

```python
num_rows = int(input("enter number of rows: "))
num_cols = int(input("enter number of columns: "))
my_matrix = []
for i in range(num_rows):
    row_vals = []
    for j in range(num_cols):
        val = int(input(f"enter element [{i}][{j}]: "))
        row_vals.append(val)
    my_matrix.append(row_vals)

print("original matrix")
for i in range(num_rows):
    print(my_matrix[i])

transposed_matrix = []
for i in range(num_cols):
    row_vals = []
    for j in range(num_rows):
        row_vals.append(my_matrix[j][i])
    transposed_matrix.append(row_vals)

print("transpose matrix")
for i in range(num_cols):
    print(transposed_matrix[i])
```
